```python
## ✅ 1. QA Difficulty Distribution (Bar Chart with Labels INSIDE)
import matplotlib.pyplot as plt

labels = ['Easy', 'Medium', 'Hard']
values = [30, 45, 25]

plt.figure()
bars = plt.bar(labels, values)

# Add labels inside bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height/2,
        f'{height}%',
        ha='center',
        va='center',
        fontsize=10,
        fontweight='bold'
    )

plt.xlabel("Difficulty Level")
plt.ylabel("Percentage")
plt.title("QA Difficulty Distribution")
plt.ylim(0, 60)

plt.show()
```

```python
## ✅ 2. Answer Type Distribution (Pie Chart with Labels INSIDE)
import matplotlib.pyplot as plt

labels = ['Factual', 'Quantitative', 'Analytical']
sizes = [40, 35, 25]

plt.figure()

wedges, texts, autotexts = plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.7  # moves % inside
)

# Style text
for autotext in autotexts:
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

plt.title("Answer Type Distribution")

plt.show()
```

```python
## ✅ 3. Chunk Count Graph (with labels inside bars)

import matplotlib.pyplot as plt

reports = ['2021-22', '2022-23', '2023-24', '2024-25']
chunks_before = [460, 433, 569, 616]
chunks_after = [460, 433, 569, 616]

x = range(len(reports))

plt.figure()
bars1 = plt.bar(x, chunks_before)
bars2 = plt.bar(x, chunks_after, bottom=chunks_before)

# Labels inside bars
for i in range(len(bars1)):
    plt.text(x[i], chunks_before[i]/2, str(chunks_before[i]),
             ha='center', va='center', fontsize=9, fontweight='bold')

    plt.text(x[i], chunks_before[i] + chunks_after[i]/2, str(chunks_after[i]),
             ha='center', va='center', fontsize=9, fontweight='bold')

plt.xticks(x, reports)
plt.xlabel("Reports")
plt.ylabel("Chunks")
plt.title("Chunk Count Distribution")

plt.show()
```

```python
## ✅ 4. Top Keywords (Horizontal Bar with Labels INSIDE)

import matplotlib.pyplot as plt

keywords = ['company', 'crore', 'year', 'bpcl', 'gas']
values = [416, 415, 413, 369, 353]

plt.figure()
bars = plt.barh(keywords, values)

# Labels inside bars
for bar in bars:
    width = bar.get_width()
    plt.text(
        width/2,
        bar.get_y() + bar.get_height()/2,
        str(width),
        ha='center',
        va='center',
        fontsize=9,
        fontweight='bold'
    )

plt.xlabel("Frequency")
plt.title("Top Keywords")

plt.show()
```
